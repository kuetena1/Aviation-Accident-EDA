## *Exploratory data analysis (EDA)* 

```python
# Checking for duplicates
df.duplicated().sum()
```

```python
# are there missing values in the dataset
print(df.isnull().sum())
print()
print(df.isnull().sum()/len(df)*100)
```

```python
# let's create a boolean of missing data and visualize on the heatmap
missing_data = df.isnull()
plt.figure(figsize=(10, 8))
sns.heatmap(missing_data, cmap = 'cividis', cbar = False,)
plt.title('Heatmap of missing data')
plt.show
        
```

```python
# For this analysis we want to maintain consistency, reliability, and robustness of the results and for that, we will drop columns with more than 30% of missing data
# let's define the threshold for columns to be dropped
thershold=0.3
missing_values = df.isnull().sum()
row_count = len(df) 
drop_columns = [i for i in df.columns if missing_values[i]/row_count>thershold]
df.drop(columns = drop_columns, inplace = True)

df.shape
```

```python
# let's check for missing value
df.isnull().sum()
```

```python
#visualizing missing value
missing_df = df.isnull()
plt.figure(figsize=(15, 8))
sns.heatmap(missing_df, cmap = 'cividis', cbar = False,)
plt.title('Heatmap of missing data')
plt.show
```

```python
#for the remaining missing data we are going to handle them differently.
# For the analysis, we are replacing the missing value for categorical columns with unknown and UNK for Wather condition

col_to_fill = ['Purpose.of.flight','Report.Status','Aircraft.damage','Amateur.Built','Injury.Severity','Engine.Type','Country']

df[col_to_fill]=df[col_to_fill].fillna('UNKNOWN')
#let's fill the weather condition with UNK and replace Unk with UNK to remain consistent

df['Weather.Condition']=df['Weather.Condition'].fillna('UNK')

df['Weather.Condition']=df['Weather.Condition'].replace({'Unk': 'UNK'})

print(df['Weather.Condition'].unique())
print ()
print(df[['Weather.Condition','Purpose.of.flight','Report.Status','Aircraft.damage','Amateur.Built','Injury.Severity','Engine.Type']])

```

```python
#replacing the null value with 0 for numerical columns

columns_to_fill = ['Total.Uninjured','Total.Minor.Injuries','Total.Serious.Injuries','Total.Fatal.Injuries']

df[columns_to_fill]=df[columns_to_fill].fillna(0)
print(df[columns_to_fill])
```

```python
# will also want to drop some of the columns that are not interesting for the analysis 

columns_to_drop = ['Number.of.Engines','Publication.Date','Registration.Number']
df=df.drop(columns_to_drop, axis =1 )
print(df.shape)
```

```python
# lest's plot the data after few cleaning process.
missing_df = df.isnull()
plt.figure(figsize=(15, 8))
sns.heatmap(missing_df, cmap = 'cividis', cbar = False,)
plt.title('heatmap of missing data')
plt.show
```

```python
# let change the columns name 
df.columns = df.columns.str.replace('.' , '_')
df.columns
```

```python
#we  also want to make sure the column Injury_Severity are consistent.
     #using regular expressions
df['Injury_Severity']= df['Injury_Severity'].replace(r'Fatal\(\d+\)','Fatal', regex=True)
     #let replace unavailable with UNKNOWN
df['Injury_Severity']= df['Injury_Severity'].replace({'Unavailable':'UNKNOWN'})
df['Injury_Severity'].unique()
```

```python
df.head(10)
```

```python
# let's extract the year and month of the event
  # Converting the event date to date format using pandas to-date function 
df['Event_Date'] = pd.to_datetime(df['Event_Date'])
#year extraction
df['year']= df['Event_Date'].dt.year
#month extraction
df['month']= df['Event_Date'].dt.month_name().str[:3]
    
df.head(10)
```

## *Observation and recomendation*

### *observation*
The majority of aviation accidents occur in the United States.
 
Cessna and Piper are the leading manufacturers associated with the highest number of accidents in the dataset.

From 1982 to 2022 they have been a significant decrease in the number of aviation accidents
 
most accidents occur under Visual Meteorological Conditions this may have some explanation. pilots can experience spatial disorientation, especially during night flights, over water, or in hazy conditions. This can also be a lack of knowledge or proficiency while using the visual fly rule. From 2010 to 2022, the number of events with unknown weather conditions has a slight upward trend. this could be an increase in unrecorded or undefined weather information over the years or potential data quality issues that may need further investigation

Jun, July, and August see a high occurrence of accidents potentially related to an increase in fly travel for holidays especially since July is a peak season for fly activity.

Accidents classified as substantial damage have more nonfatal records than fatal ones, suggesting that while the aircraft sustains significant damage, the occupants often survive. This could be due to the nature of these incidents, such as hard landings or controlled crashes, which offer higher chances of survival

Destroyed damage has more fatal records than non-fatal ones. It often indicates a high-severity crash. The force and impact involved in such accidents are likely to cause severe injuries or fatalities The fewer non-fatal records associated with destroyed aircraft suggest that the survivability rate in these incidents is low
